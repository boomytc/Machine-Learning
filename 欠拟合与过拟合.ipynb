```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_dots = 20
MAX_VALUE = 1
# [0, 1] 之间创建 20 个点
x = np.linspace(0, MAX_VALUE, n_dots)
# 立方函数，加上一点随机噪声，再加上一个固定offset
y = x**3 + 0.25 * (np.random.rand(n_dots) - 0.5)
```

```python
# 设置图片大小
plt.rcParams["figure.figsize"] = (12, 10)
plt.figure() # figsize=(10, 32), dpi=80
titles = ['Under Fitting', 'Fitting', 'Over Fitting', 'completely remembered']
models = [None, None, None, None]
```

```python
for index, order in enumerate([1, 3, 9, 18]):
    plt.subplot(2, 2, index + 1)
    p = np.poly1d(np.polyfit(x, y, order))
    # 画出拟合出来的多项式所表达的曲线以及原始的点
    t = np.linspace(0, MAX_VALUE, 200)
    # plt.plot(x, y, 'ro', t, p(t), '-', t, t ** 2, 'r--')
    plt.plot(x, y, 'ro', t, p(t), '-')
    # plt.xlim(-5, 25)
    # plt.ylim(-100, 500)

    models[index] = p
    plt.title(titles[index], fontsize=12)

plt.savefig("under-fit and over-fit.png")
plt.show()
```
