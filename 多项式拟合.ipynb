```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1000)  # x值，此时表示弧度
y = np.sin(x*np.pi/180)  #函数值，转化成度
```

```python
x[:10]
```

```python
y[:10]
```

```python
# 用5次多项式拟合，可改变多项式阶数
z1 = np.polyfit(x, y, 7)
z1
```

```python
#得到多项式系数，按照阶数从高到低排列
p1 = np.poly1d(z1)

#显示多项式
print(p1)
```

```python
# 计算x中对应的拟合的结果
yvals=p1(x)
```

```python
yvals[:10]
```

```python
plt.plot(x, y, 'g--',label='original values')
plt.plot(x, yvals, 'r',label='polyfit values')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4) # 指定legend在图中的位置，类似象限的位置
plt.title('polyfitting')
plt.show()
```

```python
# 用2次多项式拟合，可改变多项式阶数
z1 = np.polyfit(x, y, 2)
print(z1)
#得到多项式系数，按照阶数从高到低排列
p1 = np.poly1d(z1)

#显示多项式
print(p1)
# 计算x中对应的拟合的结果
yvals=p1(x)
plt.plot(x, y, 'g--',label='original values')
plt.plot(x, yvals, 'r',label='polyfit values')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4) # 指定legend在图中的位置，类似象限的位置
plt.title('polyfitting')
plt.show()
```
