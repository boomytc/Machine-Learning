```python
#1.导入画图模块
import matplotlib.pyplot as plt
```

```python
#2.实现公式
def ty(k,b,p):
    x=range(0,12)
    y=[k*i+b for i in x]
    x1=(k*(p[1]-b)+p[0])/(k*k+1)
    y1=k*x1+b
    return {"line":[x,y],"tyd":[x1,y1]}
```

```python
#3.初始化参数
k=0.6
b=0
p=[5,5]
```

```python
#4.获得相关数据
data=ty(k,b,p)
```

```python
#5.初始化作图的数据
x=data['line'][0]
y=data['line'][1]
x1=data['tyd'][0]
y1=data['tyd'][1]
```

```python
#6.作图
plt.figure(figsize=(6,6))
plt.plot(x,y,color='k')
plt.scatter(p[0],p[1],color='black',edgecolors='k')
plt.scatter(x1,y1,color='black',edgecolors='k')
plt.plot([p[0],x1],[p[1],y1],ls='--',c='k')
plt.xlim(0,11)
plt.ylim(0,11)
plt.show()
```
