```python
#1.导入本项目所需要的模块
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
```

```python
#2.导入数据集
iris=datasets.load_iris()
```

```python
#3.获取自变量数据
X=iris['data']
```

```python
#4.获取因变量数据
y=iris['target']
```

```python
#5.获取因变量名称
target_names = iris['target_names']
```

```python
#6.观察数据集
for m,i,target_name in zip('vo^',range(2),target_names[0:2]):
    s1=X[y==i,0]    #s1=sepal length (cm)
    sw=X[y==i,1]    #sw=sepal width (cm)
    plt.scatter(s1,sw,marker=m,label=target_name,s=30,c='k')

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
#7.关闭作图窗口
plt.close()
```

```python
#8.获取数据
X=X[(y==1) | (y==0),0:2]
y=y[(y==1) | (y==0)]
```

```python
#9.创建模型变量
lda=LinearDiscriminantAnalysis(n_components=1)
```

```python
#10.训练数据
ld = lda.fit(X,y)
```

```python
#11.将模型应用到原矩阵上
X_t = ld.transform(X)
```

```python
#12.转换y的结构
y_t = np.zeros(X_t.shape)
```

```python
#13.作压缩后的图像
for m,i,target_name in zip ('ov^',range(2),target_names[0:2]):
    plt.scatter(X_t[y==i],y_t[y==i],marker=m,label=target_name,s=30,c='k')

plt.legend()
plt.show()
```

```python
#14.关闭窗口
plt.close()
```

```python
#15.分割训练集和测试集
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#16.创建线性判别对象
lda = LinearDiscriminantAnalysis(n_components=1)
```

```python
#17.训练模型
ld = lda.fit(X_train,y_train)
```

```python
#18.模型预测
pre=ld.predict(X_test)
```

```python
#19.查看预测结果
list(zip(pre,y_test,pre==y_test))
```

```python
#20.查看准确率
ld.score(X_test,y_test)
```
